```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv(r"C:\Users\User\Downloads\Anova.csv")
```

```python
df
```

```python
anova1=ols('Salary~(Edu_Back)',data=df).fit()
anova=sm.stats.anova_lm(anova1,type=2)
print(anova)
```

```python
#CONCLUTION MADE
in the hypothesis we fail to reject the null hypothesis meaning Educational Background has no significant impact on Salary since the p-value(0.544324)is greater than the
common significance level(0.05 or 0.01)
```

```python
anova1=ols('Salary~(Work_Exp)',data=df).fit()
anova=sm.stats.anova_lm(anova1,type=2)
print(anova)
```

```python
#CONCLUTION MADE
in the hypothesis we reject the null hypothesis meaning Work Experience has a significant impact on Salary since the p-value(0.000022)is less than the
common significance level(0.05 or 0.01)
```

```python
anova1=ols('Salary~(Edu_Back)+(Work_Exp)',data=df).fit()
anova=sm.stats.anova_lm(anova1,type=2)
print(anova)
```

```python
#CONCLUTION MADE
p-value=0.411720 which is greater than 0.05 meaning Educational Background has no significant effect on Salary ,we fail to reject the null hypothesis for Educational Background
p-value=0.000033 which is smaller than 0.05 meaning Work Experience has a significant effect on salary therefore we reject the null hypothesis
If salary is the main goal then one should focus more on Work Experience than Educatonal Background
```

```python
anova1=ols('Salary~(Edu_Back)+(Work_Exp)*(Edu_Back)',data=df).fit()
anova=sm.stats.anova_lm(anova1,type=2)
print(anova)
```

```python
#CONCLUTION MADE
p-value=0.419353 which is greater than 0.05 meaning Educational Background has no significant effect on Salary ,we fail to reject the null hypothesis for Educational Background
p-value=0.000045 which is smaller than 0.05 meaning Work Experience has a significant effect on salary therefore we reject the null hypothesis
p-value=0.700817 which is much greater than 0.05 meaning the interaction effect between Work Experience and Educational Background has no effect on Salary ,we fail to reject the null hypothesis
```
